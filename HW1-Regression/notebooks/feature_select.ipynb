## Naive feature selection

特征选择是 HW1 的重要一步，我们需要移除相关性较小的特征。首先我们仅从直觉上移除了以下部分特征，其余特征有待进一步推敲：

- id
- States (35, encoded to one-hot vectors) 
- Belief indicators
    - belief_mask_effective, belief_distancing_effective.
- Mental indicator
    - worried_catch_covid, worried_finance.

id 和州的信息显然和 tested_positive 没有明显的相关性。另外，从直觉上来说关于 Belief indicators 和 Mental indicator 并不是重点，或许我们应该更关注他们的具体行为而不是如何认为的。

```python
# 删除前35列州信息
data_without_states = data.iloc[:, 35:]
data_without_states
```

```python
# 删除包含"belief"的列
belief_columns = [col for col in data_without_states.columns if "belief" in col]
data_without_states_belief = data_without_states.drop(columns=belief_columns)
data_without_states_belief
```

```python
# 删除包含"worried"的列
worried_columns = [col for col in data_without_states.columns if "worried" in col]
data_without_states_belief_worried = data_without_states_belief.drop(columns=worried_columns)
data_without_states_belief_worried
```

## Selecting: Human or Machine?

机器自己通过深度学习去给每个 Feature 设置并更新权重，对于不那么重要的 Feature 相应的 Feature也会降低。那么我们为什么还要自己去做 Feature Selection 呢？我尝试把 Naive Select 的结果去让机器学，但是测试的结果并没有很好。所以我参考 [a86gj387](https://zhuanlan.zhihu.com/p/483652591) 的做法，计算 Feature 和 tested_positive 的相关度，只选择相关度 $>0.5$ 的部分。结果确实有了不错的提升。可能是因为训练的数据量还是太少了，机器并没有很好地学习到如何去选择 Feature。

```python
cor = data.corr()["tested_positive"]
best_cor = cor[cor>0.5]
best_cor
```

```python
best_data = data[best_cor.index]
best_data
```
